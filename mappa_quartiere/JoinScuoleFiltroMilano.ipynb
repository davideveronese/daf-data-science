```python
import pandas as pd
```

```python
affluenza = pd.read_csv('data/affluenzaScuole1617.csv')
anagrafica = pd.read_csv('data/anagrafeScuole1617.csv').set_index('CODICESCUOLA')

#datiMi = anagrafica

#filtro = pd.read_csv('data/codScuolaMilano.csv',header=None).squeeze()
```

```python
cittaScelta = 'MILANO'
anagrafica = anagrafica[anagrafica.DESCRIZIONECOMUNE == cittaScelta]
anagrafica.to_csv('anagrScuoleMilano.csv', sep=';')

```

```python
#import seaborn as sns
#sns.barplot(data = datiMi, y='FASCIAETA', x='ALUNNI', orient='h')
```

```python
# Join con affluenza:
datiJoin = anagrafica.merge(affluenza, how='inner', left_index=True, right_on='CODICESCUOLA',
                sort=False, suffixes=('_anagr', '_affl'), copy=True, validate='1:m')

datiJoin.to_csv('datiScuolaJoinedMilano.csv')
```
